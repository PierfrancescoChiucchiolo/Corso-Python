```python
#Import delle librerie
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from faker import Faker
import datetime
```

```python
## generiamo le date da inserire come indici nel dataframe
date = pd.date_range( start = "2025-01-01", end = "2025-12-31", periods = 365)

## for data in date: print(data)
```

```python
## genera il numero di visite giornaliere con una gaussiana
## media di 1200 pazienti, deviazione standard di 900, nell'arco di 365 giorni

def visite_odierne(base, deviazione):

    visite = np.int64(np.random.normal(loc = base, scale = deviazione))
    return visite

print( visite_odierne(120, 90) )
```

```python
## creo una funzione che giorno dopo giorno mi crea un certo numero di nomi falsi con faker, usando una serie di date generate altrove

def fake_people(people):

    lista = []
    for i in range(0, people):
        lista.append(Faker().name())

    serie = pd.Series(lista)
    return serie

print(fake_people(10))
```

```python
def genera_malattie(malattie, numero):

    lista = []
    for i in range(0, numero):
        lista.append( np.random.choice(malattie) )

    serie = pd.Series(lista)
    return serie

lista_malattie = ["Ossa", "Cuore", "Testa"]
print(genera_malattie( lista_malattie , 10))
```

```python
## generiamo il dataset

lista_malattie = ["Ossa", "Cuore", "Testa"]

print(date)
dati = {}
for giorno in date:

    odierno = {}
    visite = visite_odierne(600, 450)
    persone = fake_people(visite)
    malattie = genera_malattie(lista_malattie, visite)
    temp = {"giorno" : giorno, "pazienti" : persone, "malattie" : malattie}
    print(temp)
    odierno.update( temp )
dati.update(odierno)
    
df = pd.DataFrame(dati)
```

```python
## numero medio visitatori medio al mese / deviazione standard / patologia più e meno trovata

mesi = ["Gennaio", "Febbraio", "Marzo", "Aprile", "Maggio", "Giugno", "Luglio", "Agosto", "Settembre", "Ottobre", "Novembre", "Dicembre"]
giorni_mesi = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

## rubato da https://stackoverflow.com/questions/11621740/how-to-determine-whether-a-year-is-a-leap-year
def is_leap_year(year):
    """Determine whether a year is a leap year."""
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)
## se leapyear allora febbraio +1


attuale = 1
for mese in mesi:

    mese_corrente = df.loc[str(datetime.date(year = 2025, month = attuale, day = 1)) : str(datetime.date(year = 2025, month = attuale, day = giorni_mesi[attuale-1]))]
    media = mese_corrente["pazienti"].count()

    attuale += 1

    print("Nel mese di " + mese + " la clinica ha avuto un totale di " + str(media) + " pazienti al giorno")
```
